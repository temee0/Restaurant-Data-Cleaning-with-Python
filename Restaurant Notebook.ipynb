### Handling NaN(Null) values in the dataset

```python
Data.isna().sum() #Loads the sum of null(NaN) values in each column 
```

**Handling "Rate" column null values**

```python
#The Rate column is string value so we will convert to float
Data['Rate'] = pd.to_numeric(Data['Rate'].str.split('/').str[0], errors='coerce')
```

```python
"""
Fill missing values in the 'Rate' column using the median rating of each restaurant 
based on the restaurant name in the 'Name' column
"""
Data['Rate'] = Data.groupby('Name')['Rate'].transform(lambda x: x.fillna(x.median()))
```

```python
"""
Attempted to fill missing 'Rate' values using the median rating per restaurant, but it did not work 
because many missing values belong to restaurants with no existing ratings (all NaN).
Since these entries do not contribute meaningful information for EDA, they will be dropped from the dataset.
"""
Data.dropna(subset=["Rate"], inplace=True)
```

**Handling "Dish_Liked" column null values**

```python
"""
Fill missing values in the 'Dish_Liked' column using the mode value of each restaurant 
based on the restaurant name in the 'Name' column. if the restuarant happens to not have a dishlike value, 
we will fill as "Unknown"
"""
Data['Dish_Liked'] = Data.groupby('Name')['Dish_Liked'].transform(
    lambda x: x.fillna(x.mode()[0] if not x.mode().empty else 'Unknown')
)
```

**Handling "Approx_Cost(For Two People)" column null values**

```python
#converting the data type to int
Data['Approx_Cost(For Two People)'] = Data['Approx_Cost(For Two People)'].astype(str).str.replace(',', '') #remove commas in our values
Data['Approx_Cost(For Two People)'] = pd.to_numeric(Data['Approx_Cost(For Two People)'], errors='coerce').astype('Int64')
```

```python
"""
Fill missing values in this column using the median value of each restaurant 
based on the restaurant name in the 'Name' column
"""
Data['Approx_Cost(For Two People)'] = Data.groupby('Name')['Approx_Cost(For Two People)'].transform(lambda x: x.fillna(x.median()))
```

```python
#check null values in Approx_Cost(For Two People) column
Data['Approx_Cost(For Two People)'].isna().sum()
```

```python
"""
There's still some NaN Values in the Approx_Cost(For Two People) column. 
I will be filling them using the general median
"""
Data['Approx_Cost(For Two People)'] = Data['Approx_Cost(For Two People)'].fillna(
    Data['Approx_Cost(For Two People)'].median())
```

```python
Data.isna().sum() #loading the sum of the null values, to confirm that i have succesfully removed all null values
```

### Cleaning individual columns

```python
#Cleaning values in the "Name" column with weird characters
Data['Name'][10] #This is an example of such weird value
```

```python
Data['Name'] = Data['Name'].str.replace(r'[^\x00-\x7F]+', '', regex=True)
```

```python
Data['Name'][10] #checking if this solved the problem with the characters
```

```python
#cleaning extra spaces in the "Name" and "Dish_Liked" column
Data[['Name', 'Dish_Liked']] = Data[['Name', 'Dish_Liked']].apply(lambda col: col.str.strip())
```

```python
#change column name from "rate" to "ratings"
Data.rename(columns = {"Rate": "Ratings"}, inplace=True)
```

```python
#resetting the index because we dropped some rows earlier 
Data.reset_index(inplace=True)
```

```python
#dropping the old index
Data.drop(columns= ['index'], inplace= True)
```
